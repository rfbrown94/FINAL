```python
%%R

let_num = paste(paste(LETTERS, '->'), paste('letter', 1:length(LETTERS)))
print(let_num)

voyels = LETTERS[c(1,5,9,15,21)]
print(voyels)

consonants = LETTERS[-c(1,5,9,15,21)]
print(consonants)
```

```python
%%R

create.names = function(n){
    create.letters = function(choice, size){
        sample(choice, size=size, prob=rep(1/length(choice), length(choice)), replace=T)
    }
    con = create.letters(consonants, n)
    v1 = create.letters(voyels, n)
    v2 = create.letters(voyels, n)
    return(paste(con, v1, v2, sep=""))
}

first = create.names(1000)
mid = create.names(1000)
last = create.names(1000)

names = paste(first, mid, last)


print(head(names)); print(tail(names))
```

```python
%%R

country = c('US', 'Italy', 'France', 'Brazil', 'China', 'Sweden', 'Japan')
sex = c('M', 'F')
occupation = c('Programmer', 'Teacher', 'CEO', 'Manager', 'Researcher', 'Unemployed', 'Lawyer', 'Student')

create.df = function(n){
    names = create.names(n)
    create.whatever = function(choice, n){
        sample(choice, n, prob=rep(1/length(choice), length(choice)), replace=T)
        }
    country = create.whatever(country, n)
    sex = create.whatever(sex, n)
    occupation = create.whatever(occupation, n)
    return(data.frame(names, country, sex, occupation))
    }

df = create.df(1000)
print(head(df))

write.csv(df, 'simulated_data.csv', row.names=F)
    
```

```python
%%R

create.grades = function(n, score.max){
    mean = sample(40:70, 1)
    sd = sample(10:30, 1)
    vec = round(rnorm(n, mean, sd))
    vec[vec>score.max] = score.max
    vec[vec<0] = 0
    return(vec)
}
grades = create.grades(100, 100)
print(head(grades))
hist(grades)
```

```python
%%R

create.categories = function(n, categ){
    create.prob = function(n){
        prob = runif(n, 0, 1)
        prob = prob/sum(prob)
        return(prob)
        }
    prob = create.prob(length(categ))
    cate_vec = sample(categ, n, prob=prob, replace=T)
    return(cate_vec)
    }
```
