##Simulate and visualizing clusters and relations##
###Simulating explanatory variables: X###

```python
%%R

clean.range = function(x, range){
    x[x<range[1]] = range[1]
    x[x>range[2]] = range[2]
    return(x)
}

simulate.x = function(var.size, var.mean=0, var.sd=1, range=NULL){
    x = round(rnorm(var.size, var.mean, var.sd))
    if (!is.null(range)){
        x = clean.range(x, range)
        }
    return(x)
}


```

```python
%%R

simulate.y = function(x.var, alpha, beta, e.mean=0, e.sd=10, range=NULL){
    error = rnorm(length(x.var), e.mean, e.sd)
    y = alpha+x.var*beta+error
    if (!is.null(range)){
        y = clean.range(y, range)
    }
    return(y)
}
```

```python
%%R
x.variable = simulate.x(var.size=1000, var.mean=0, var.sd=100)
y.variable = simulate.y(x.variable, 2, 0.5)

```

```python
%%R

reg = lm(y.variable ~ x.variable)
plot(x.variable, y.variable)
abline(reg)
```

Benoit's simulation of Y-variable, using matrix multiplications.

```python
%%R

simulate.Y = function(X.variables,      #
                      alpha,            #
                      beta,             #
                      residual.sd=1,    #
                      residual.mean=0,  #
                      var.range=NULL    #
                     )
    {
        X = as.matrix(X.variables)
        residuals = rnorm(dim(X)[1], mean=residual.mean, sd=residual.sd)
        e = matrix(residuals, ncol=1)
        v = matrix(beta, ncol=1)
    
        Y = alpha + X %*% v + e
    
        if(!is.null(var.range)) Y = clean.range(Y, var.range)
        return(as.numeric(Y))
    }
```

###Simulating clusters###

```python
%%R
##cluster info to be past to x and y function
sizes.def = c(30, 40, 50)
mean.def = c(-50, 100, 50)
sd.def = c(10, 10, 10)
alpha.def = c(20, 10, 30)
betas.def = c(0.9, 1.2, -0.9)
```

```python
%%R

cluster.info = function(cl.sizes=sizes.def,
                        cl.means = mean.def,
                        cl.sd = sd.def,
                        cl.alphas = alpha.def,
                        cl.betas = betas.def){
    cl.list = list()
    cl.number = length(cl.sizes)
    sample.size = sum(cl.sizes)
    
    for (i in 1:cl.number){
        cl.name = paste('cluster', i, sep="")
        cl.list[[cl.name]] = c(size=cl.sizes[i],    
                                   mean=cl.means[i],    
                                   sd=cl.sd[i],         
                                   alpha=cl.alphas[i],  
                                   beta=cl.betas[i])
        }
    return(t(data.frame(cl.list)))
    }
```

```python
%%R

cluster.default = cluster.info()
print(cluster.default)
```

```python
%%R

simulate.clusters = function(cl.info){
    X = simulate.x(cl.info['size'], cl.info['mean'], cl.info['sd'])
    Y = simulate.y(X, cl.info['alpha'], cl.info['beta'])
    return(data.frame(X=X, Y=Y))
    }
```

```python
%%R

simulate.clusterdata = function(cluster.info=cluster.default){
    m = nrow(cluster.info)
    data = data.frame()
    for (i in 1:m){
        df = simulate.clusters(cluster.info[i,]) #takes ith line in cluster.default
        df$Cluster = factor(rep(i, nrow(df)))
        data = rbind(data, df)
    }
    shuffle = sample(nrow(data))
    data = data[shuffle, ]
    return(data)
}
```

```python
%%R

data = simulate.clusterdata()
print(head(data))
```

```python
%%R 
print(cluster.default[1,])
```

###True clusters and true relations

```python
%%R

my_colors = c('red', 'blue', 'green')
colors = my_colors[data$Cluster]

plot(data$X, data$Y, col=colors)

plot.truelines = function(){
    for (cluster in 1:3){
        intercept = cluster.default[cluster, 'alpha']
        coefficient = cluster.default[cluster, 'beta']
        color = my_colors[cluster]
        abline(intercept, coefficient, col=color)
        }
    }

plot.truelines()
```

##Computing and visualizing a regression##


```python
%%R

model = lm(data$Y ~ data$X)
print(model)
```

Extracting information from the model
   
    abline(model)
    fitted(model)
    resid(model)

```python
%%R

plot(data$X, data$Y, col=colors)

abline(model, col='orange')
```

```python
%%R

plot(data$X, data$Y, col=colors)

points(data$X, fitted(model), col='orange')
segments(data$X, fitted(model), data$X, data$Y)
```

```python
%%R

error = sum(resid(model)^2)
print(error)
```

```python
%%R

distance = dist(data[c(1,2)])
htree = hclust(distance)
plot(htree)
```

```python
%%R

clusters = cutree(htree, 3)
print(head(clusters))
```

```python
%%R

A = clusters[clusters==1]
B = clusters[clusters==2]
C = clusters[clusters==3]

print(head(A))
print(names(A))
```

```python
%%R

data.A = data[names(A), ]
data.B = data[names(B), ]
data.C = data[names(C), ]

model.A = lm(data.A$Y ~ data.A$X)
model.B = lm(data.B$Y ~ data.B$X)
model.C = lm(data.C$Y ~ data.C$X)
```

```python
%%R

my_colors = c('red', 'blue', 'green')
colors = my_colors[clusters]

plot(data$X, data$Y, type='n')

plot.cluster = function(cluster, model, color){
    points(cluster$X, cluster$Y, col=color)
    segments(cluster$X, fitted(model), cluster$X, cluster$Y)
    lines(cluster$X, fitted(model), col=color)
    abline(model, col=color)
    }
plot.cluster(data.A, model.A, 'red')
plot.cluster(data.B, model.B, 'blue')
plot.cluster(data.C, model.C, 'green')
```

```python
%%R

par(mfrow=c(1,3))

plot(data.A$X, data.A$Y, type='n')
plot.cluster(data.A, model.A, 'red')

plot(data.B$X, data.B$Y, type='n')
plot.cluster(data.B, model.B, 'blue') 

plot(data.C$X, data.C$Y, type='n')
plot.cluster(data.C, model.C, 'green')
```

##Interpreting and prediction with a regression (multivariable)##

```python
%%R

var.num = 10
sample.size = 200
var.mean = 100
var.sd = 200

x = replicate(var.num, rnorm(sample.size, var.mean, var.sd))

colnames(x) = paste('x', 1:var.num, sep="")

print(head(x))
```

```python
%%R

alpha = 10
beta = rnorm(var.num)
error = rep(0, var.num)

y = simulate.Y(x, alpha, beta)

par(mfrow=c(1,2))
hist(y)
boxplot(y)
```

```python
%%R

df = data.frame(x, y=y)
print(head(df))

ind = df$y>2000

df.outliers = df[ind, ]
print(df.outliers)
```

```python
%%R

model = lm(y ~ ., data=df)
stats = summary(model)
print(stats)
```

```python
%%R

a = model$coefficients[1]
b = model$coefficients['x1']
print(c(a,b))
```

```python
%%R

plot(df$x1, df$y)
abline(a, b)
```
