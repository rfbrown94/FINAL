##Factors##
The class factor is used in R to represent **categorical variables**. The class constructor

    factor(X, levels=values, ordered=TRUE)

takes

* a vector X containing the values of our categorical variable

* a vector levels containing the category names

* a Boolean argument ordered specifying whether the categories are ordered or not (e.g. 'Cold', 'Warm', and 'Hot' are ordered, but 'Male' and 'Female' are not)

and returns a **factor** representing the categorical variable.

```python
%%R

t = factor(temp, levels=c('Cold', 'Warm', 'Hot'), ordered=T)
print(head(t))
print(table(t))
```

```python
%%R

barplot(table(t))
```

##The paste function##

Simulating a data frame will most probably involve generating character vectors for row or column names. For that task, use the function

    paste(v1, v2, v3, etc., sep='_')
    
**RECYCLING**: If the vectors v1,v2, and v3 are not of the same size, the vectors of smallest lenght will be recycled into vectors length equal to the maximal length by repeating the elements in a cycle. For instance: Before concatenation of the two vectors

     (1, 2, 3, 4, 5)
     (x, y)

paste will first recycle the smallest vector and then preform an element-wise concatenation of the vectors

     (1, 2, 3, 4, 5)
     (x, y, x, y, x)


```python
%%R
students = paste('Student', 1:100, sep=" ")
print(tail(students))
```

##The built-in vectors letters and LETTERS##

R has two convenient built-in character vectors

    letters
    LETTERS

containing the alphabet letters in lower-case for the former and in upper-case for the later.

One can use them to generate dummy names.

```python
%%R

letter_order = paste(LETTERS, 1:26, sep=" -> letter ")
print(letter_order)
```

```python
%%R

voyels = LETTERS[c(1, 5, 9, 15, 21)]
consonants = LETTERS[-c(1, 5, 9, 21, 24)]

print(voyels); print(head(consonants))
```

```python
%%R

generate.voyel = function() return(voyels[sample(1:length(voyels), 1)])
generate.consonant = function() return(consonants[sample(1:length(consonants), 1)])
generate.one = function() return(paste(generate.consonant(), paste(generate.voyel(), generate.voyel(), sep=""), sep=""))

create.names = function(n){
    names = c()
    for (i in 1:n){
        name = paste(generate.one(), generate.one(), generate.one(), sep=" ")
        names = c(names, name)
    }
    return(names)
}
      
```

```python
%%R

names = create.names(10)
print(names)
    
```

##runif function##

The function

    runif(n, min, max)
    
return a numeric vector of n real numbers uniformally choosen between min and max.

```python
%%R

create.prob = function(n){
    prob = runif(n, min=0, max=1)
    return(prob/sum(prob))
}

create.categories = function(n, categ){
    freq = create.prob(length(categ))
    values = sample(categ, n, replace=T, prob=freq)
    return(values)
    }


```
